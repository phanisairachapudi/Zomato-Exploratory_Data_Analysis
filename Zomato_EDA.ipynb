```python
# import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Get the zomato dataset

df_zomato = pd.read_csv('zomato.csv', encoding = 'latin-1')
```

```python
# view what are all columns are there in dataset

df_zomato.columns
```

Based on above information. Cuisines has 9 missing values

Now combine our zomato dataset with country-code dataset. We're doing because we've country code in zomato dataset. But, it was missing country name.

```python
# Get the country-code dataset

df_country = pd.read_excel('country-code.xlsx')
```

```python
# view first 5 records

df_country.head()
```

```python
# view any null values are there

df_country.isnull().sum()
```

Using country code column in zomato and country datasets. we are comibing these two datasets using pandas merge method.

```python
df_final = pd.merge(df_zomato, df_country, on = 'Country Code', how = 'left')
```

```python
# view first 5 records

df_final.head()
```

```python
# view what are all columns are there in dataset after merged

df_final.columns
```

```python
# check how many records based on countries

df_final.Country.value_counts()
```

Plotting a graph for showing which top 3 countries uses zomato most

```python
country_names = df_final.Country.value_counts().index
```

```python
country_values = df_final.Country.value_counts().values
```

```python
# Pie Charts

plt.pie(country_values[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

Zomato has maximum transactions are from India with 94.39%. 
United States is 4.73%.
United Kingdom is 0.87%.

Using final dataset find out best restaurants based on Aggregate Ratings

```python
df_best = df_final.sort_values(by='Aggregate rating', ascending = False).copy()

df_best.loc[:,["Restaurant Name", "City", "Address", "Cuisines", "Votes", "Aggregate rating"]]
```

Top 10 resturants based on ratings

```python
def_best_10 = df_best.loc[:,["Restaurant Name", "City", "Address", "Cuisines", "Votes", "Aggregate rating"]].head(10)

def_best_10
```

Let's find out best dishes(Cuisines) based on votes

Let's take one city to find details about that city. For example: Noida

```python
df_final.City.value_counts().head(20)
```

```python
df_city = df_final[(df_final.City == 'New Delhi')]
```

```python
df_city.head(10)
```

```python
# Find most preferred dishes or top 10 dishes in noida

plt.figure(figsize = (20,10))
sns.barplot(y = df_city.Cuisines.value_counts()[:10,].index, x = df_city.Cuisines.value_counts()[:10,], data = df_city)
```

In New Delhi city, most people eat North Indian, Chinese, Fast Food
